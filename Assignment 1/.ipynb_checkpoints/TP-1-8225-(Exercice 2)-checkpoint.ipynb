```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split

# We import the data
digits = datasets.load_digits ()
X= digits.data
y = digits.target

# Boolean for adding new dimensions
add_values = True
new_dim = 0
nb_dim = 8

if add_values:
    new_X = np.zeros((X.shape[0],X.shape[1]+nb_dim))

    for i in range(X.shape[0]):
        for j in range(1,nb_dim):
            new_dim = np.random.uniform(np.min(X),np.max(X),1)
            new_X[i,X.shape[1]+j]=new_dim
    new_X[:,:X.shape[1]]=X[:,:]
    
    X = new_X
    
# Column of ones added to X    
X = np.c_[np.ones(X.shape[0]),X]
    
```

```python
# Boolean for using regularization (set to yes if so)
reg = True
```

```python
# The functions used in our model

def softmax(x):
    return np.exp(x) / np.sum(np.exp(x),axis=0)

def get_accuracy(X, y, W):
    sum = 0
    for v in range(0, len(X)):
        y_pred = softmax(np.dot(W, X[v]))
        sum = sum + np.vdot(y[v], y_pred)
    return sum / len(X) * 100 

def get_grads(y,y_pred,x):
    return 2/minibatch_size*(np.dot(y_pred.T,x)-np.dot(y.T,x))
    
    
def get_loss(y, y_pred):
    res = 0
    for i in range(y_pred.shape[0]):
        for j in range(y_pred.shape[1]):
            if y[i,j]==1:
                res = res + np.log(y_pred[i,j])
    res = (-1)*res/(y.shape[0])
    return res
```

```python
# Encoding our target vector and splitting data into train, validation and test sets

y_one_hot = np.zeros((y.shape[0], len(np.unique(y))))
y_one_hot[np.arange(y.shape[0]) , y] = 1 
X_train, X_test, y_train, y_test = train_test_split(X, y_one_hot, test_size=0.3, random_state=42)
X_test, X_validation, y_test, y_validation = train_test_split(X_test, y_test, test_size=0.5,random_state=42)
```

```python
# Initialization of the model's hyperparameters

#minibatch_sizes = [10,50,100,500,1000]
#learning_rates = [0.1,0.01,0.001,0.0001]
minibatch_sizes = [20]
learning_rates = [0.001]
results = []
liste_lambdas_1 = [1, 2, 5, 10, 20, 100, 1000]
liste_lambdas_2 = [1, 2, 5, 10, 20, 100, 1000]
#liste_lambdas_1 = [1]
#liste_lambdas_2 = [1]
#lambda_1 = 1
#lambda_2 = 50
nb_epochs = 50
liste_alphas = []
```

```python
# First loop on lambda_1
for l in liste_lambdas_1:
    
    lambda_1 = l
    
# Second loop on lambda_2
    for L in liste_lambdas_2:
        
        lambda_2 = L
        alpha = lambda_2/(lambda_1+lambda_2)
        liste_alphas.append(alpha)
        
# Third loop on minibatch_size
        for size_minibatch in minibatch_sizes:

            minibatch_size = size_minibatch

# Fourth loop on the learning rate
            for learning_rate in learning_rates:

                list_parameters = []

                lr = learning_rate


                # We initialize our variables to keep track of our model's performance
                best_W = None
                accuracies_val = []
                accuracies_train = []
                best_accuracy = 0
                losses_train = []
                losses_val = []

                # We initialize the weights matrix with a normal distribution
                W = np.random.normal(0,0.01,(len(np.unique(y)), X.shape[1]-1)) 
                # We create theta (=concatenation of W and vector of ones)
                theta = np.c_[W,np.ones(W.shape[0])]

                for epoch in range(nb_epochs):

                    loss_train = 0
                    loss_val = 0
                    accuracy_val = 0
                    accuracy_train = 0

                    for i in range (0, X_train.shape[0], minibatch_size):

                        # We select mini-batch sized sets of data according to the value of i
                        if (i+minibatch_size)>X_train.shape[0]:
                            xi = X_train[i:X_train.shape[0],:]
                            yi = y_train[i:y_train.shape[0],:]
                        else : 
                            xi = X_train[i:i+minibatch_size,:]
                            yi = y_train[i:i+minibatch_size,:]

                            
                        # We train our model on minibatches
                        y_pred = softmax(np.dot(theta,xi.T)).T
                        grad = get_grads(yi, y_pred, xi)
                        
                        # Update the gradient using regularization or not 
                        if reg : 
                            theta = theta-lr*(grad+(1-alpha)*np.sign(theta)+2*alpha*np.sum(theta[1:]))
                            # We take np.sum(theta[1:]) because we don't consider the column of ones
                        else : 
                            theta = theta-lr*grad

                    # Evaluating our model on the whole train set
                    y_pred = softmax(np.dot(theta, X_train.T)).T
                    if reg:
                        loss_train = get_loss(y_train, y_pred)+alpha*np.sum(theta)+(1-alpha)*np.sum(theta[1:]**2)
                    else :
                        loss_train = get_loss(y_train, y_pred)
                    losses_train.append(loss_train)      
                    accuracy_train = get_accuracy(X_train, y_train, theta)
                    accuracies_train.append(accuracy_train)


                    y_pred = softmax(np.dot(theta, X_validation.T)).T
                    loss_val = get_loss(y_validation, y_pred)
                    losses_val.append(loss_val) 
                    accuracy_val = get_accuracy(X_validation, y_validation, theta)
                    accuracies_val.append(accuracy_val)

                    # We save the weights giving the best accuracy (to counter overfitting)
                    if accuracy_val > best_accuracy: 

                        best_accuracy = accuracy_val
                        best_W = theta      
                # Testing the model on the test set
                accuracy_on_unseen_data = get_accuracy(X_test, y_test, best_W)


                # Saving parameters and history of the model
                list_parameters.append(size_minibatch)
                list_parameters.append(learning_rate)
                list_parameters.append(accuracy_on_unseen_data)
                #list_parameters.append(losses_train)
                #list_parameters.append(losses_val)
                #list_parameters.append(accuracies_train)
                #list_parameters.append(accuracies_val)
                list_parameters.append(np.mean(best_W[:,64:72]))
                list_parameters.append(np.var(best_W[:,64:72]))
                list_parameters.append(np.mean(best_W[:,:64]))
                list_parameters.append(np.var(best_W[:,:64]))
                if reg :
                    list_parameters.append(lambda_2)
                    list_parameters.append(lambda_1)

                results.append(list_parameters)
```

```python
print('Displaying the different parameters combinations :\n')
print('-------')
for i in range(len(results)):
    print('Parameters used :\n')
    print('minibatch size :',results[i][0])
    print('learning rate :',results[i][1])
    if reg:
        print('valeur de lambda_1 :', results[i][-1])
        print('valeur de lambda_2 :', results[i][7])
    print('moyenne poids non_ex :', results[i][3])
    print('variance poids non_ex :', results[i][4])
    print('moyenne poids expl :', results[i][5])
    print('variance poids expl :', results[i][6])
    print('Obtained accuracy on validation set :', results[i][2],'%')
    print('-------')
```

```python
# We look for the set of parameters giving the best accuracy results in our results list

best_config_index = 0
maximum_acc = 0
for i in range(len(results)):
    if results[i][2]>maximum_acc:
        best_config_index = i
        maximum_acc = results[i][2]
```
